#### 01) Qual é a idade, em anos, do titular mais jovem?

```python
# Conferindo os dados
dias = df['DAYS_BIRTH']*(-1)
dias
```

```python
# Convertendo o valor dias em anos
anos = dias/365
anos
```

```python
# Conferindo os 5 mais novos e os 5 mais velhos
anos.sort_values(ascending=True)
```

```python
print(f'O titular mais jovem possui: {int(anos.min())} anos')
```

#### 02) Qual a média do valor de anuidade dos residentes em apartamentos maiores de 50 anos?

```python
df1 = df[['AMT_ANNUITY','DAYS_BIRTH']]
df1
```

```python
selecao = (df1['DAYS_BIRTH']*(-1)) > (50*365)
```

```python
s1 = df1[selecao]
s1
```

```python
s2 = s1.sort_values(by=['AMT_ANNUITY'])
s2
```

```python
s2 = s1[s1['AMT_ANNUITY'].notnull()]
s2
```

```python
s2.iloc[0:1, :].mean()
```

```python
# Resposta final:
# a média do valor de anuidade dos residentes em apartamentos maiores de 50 anos é de:
# R$ 6.750,00
```

#### 05) Qual é a diferença entre a média de valor de crédito e a média de valor de anuidade quando filtramos pelos trabalhadores do tipo de organização Military?

```python
# Conferindo os dados
tipo_organizacao = df['ORGANIZATION_TYPE']
tipo_organizacao
```

```python
# Excluindo duplicatas
tipo_organizacao = tipo_organizacao.drop_duplicates()
tipo_organizacao
```

```python
# Conferindo quantidade de tipos de organização
tipo_organizacao.shape[0]
```

```python
tipo_organizacao = pd.DataFrame(tipo_organizacao)
```

```python
tipo_organizacao = tipo_organizacao.sort_values(by=['ORGANIZATION_TYPE'])
```

```python
# Ordenando o index
tipo_organizacao.index = range(tipo_organizacao.shape[0])
tipo_organizacao
```

```python
# Selecionando colunas
s3 = df[['ORGANIZATION_TYPE', 'AMT_CREDIT', 'AMT_ANNUITY']]
s3
```

```python
# Retirando os nulos
selecao = s3[s3['AMT_ANNUITY'].notnull()]
selecao
```

```python
# filtrando militares
militares = 'Military'
filtro = (selecao['ORGANIZATION_TYPE'] == militares)
filtro
s4 = selecao[filtro]
s4
```

```python
# tirando a média
media_credito =  s4['AMT_CREDIT'].mean()
media_anuidade = s4['AMT_ANNUITY'].mean()
```

```python
# RESPOSTA FINAL
media_anuidade.round(2)
```

```python
media_credito
```

```python
diferenca = media_credito - media_anuidade
diferenca
```

```python
print(f'A diferença é de: R$ {diferenca:.2f}')
```
